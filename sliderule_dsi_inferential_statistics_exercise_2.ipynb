```python
data.call[data.race == 'b']
# alternative way of subsetting: data[data.race=='w'].call
# call values for race == b, list of values for which race == b
```

```python
call_b = data.call[data.race =='b']
call_w = data.call[data.race =='w']

b_mu = np.mean(call_b) # call back rate, success rate
w_mu = np.mean(call_w)

b_tot = len(call_b)
w_tot = len(call_w)

print(b_mu, w_mu)
# 2:
# null: no discrimination
# alt: discrimination
```

```python
t, p= s.ttest_ind(a=call_b, b=call_w, equal_var=False)
print(t,p)
# reject the null
```

```python
# 3:
se_b = s.sem(call_b)
se_w = s.sem(call_w)
print(se_b,se_w)
# standard errors
```

```python
ci_b = s.t.interval(alpha = .95, df = b_tot-1, loc = b_mu, scale=s.sem(call_b))
ci_w = s.t.interval(alpha = .95, df = w_tot-1, loc = w_mu,scale=s.sem(call_w))
print(ci_b,ci_w)
```
